# Regression Analysis

```python
import glob
import statsmodels
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
data_path = glob.glob('data/extracted/**/*.csv', recursive=True)[0]
df_original = pd.read_csv(data_path, low_memory=False)
```

```python
df = df_original.copy()
df["IsMale"] = df["Sex"].map(lambda sex: 1 if sex == "M" else 0)
df["IsTested"] = df["Tested"].map(lambda tested: 1 if tested else 0)
df
```

## Lift Level

```python
import statsmodels.formula.api as sm
FORMULA = "TotalKg ~ Age + IsMale + BodyweightKg + IsTested"
result = sm.ols(formula=FORMULA, data=df).fit()
result.summary()
```

## Lifter Level

```python
df_lifters = df.groupby("Name").max(["TotalKg"])
df_lifters["TotalKg"]
```

```python
FORMULA = "TotalKg ~ Age + IsMale + BodyweightKg + IsTested"
result = sm.ols(formula=FORMULA, data=df_lifters).fit()
result.summary()
```
